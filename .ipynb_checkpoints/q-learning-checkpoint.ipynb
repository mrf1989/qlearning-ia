### Clase Board (Tablero)

Se ha creado una clase que representa al tablero (Board) con los siguientes atributos:

- n: número de filas del tablero.
- m: número de columnas del tablero.
- initial_state: casilla desde la que se desea partir.
- goal_state: casilla objetivo que se pretende alcalzar.
- map_squares: diccionario que relaciona una coordenada del tablero (clave), mediante una tupla, con un número de casilla (valor).
- map_states: diccionario que relaciona un número de casilla (clave), mediante un entero único de ``0 a (n*m)-1``, con una coordenada del tablero, mediante una tupla (valor).
- R: matriz de recompensas que se utiliza para llevar a cabo los entrenamientos.


````python
class Board():
    def __init__(self, n, m, initial_state, goal_state):
        self.__n = n
        self.__m = m
        self.__initial_state = initial_state
        self.__goal_state = goal_state
        self.__map_squares = {}
        self.__map_states = {}
        self.__R = []
    
    def get_n(self):
        return self.__n
    
    def set_n(self, n):
        self.__n = n
    
    def get_m(self):
        return self.__m
    
    def set_m(self, m):
        self.__m = m

    def get_size(self):
        return self.__n * self.__m
    
    def get_R(self):
        return self.__R

    def get_map_states(self):
        return self.__map_states
    
    def get_initial_state(self):
        return self.__initial_state
    
    def set_initial_state(self, initial_state):
        self.__initial_state = initial_state
    
    def get_goal_state(self):
        return self.__goal_state
    
    def set_goal_state(self, goal_state):
        self.__goal_state = goal_state

    def init_random_board(self):
        states = sample([i for i in range(self.get_size())], self.get_size())
        k = 0

        for i in range(self.__n):
            for j in range(self.__m):
                state = states[k]
                self.__map_squares[(i, j)] = state
                self.__map_states[state] = (i, j)
                k = k + 1
    
    def is_neighbor(self, square_src, square_dst):
        distance = sqrt(power(square_dst[0] - square_src[0], 2) +
                    power(square_dst[1] - square_src[1], 2))
        return distance <= sqrt(2) and distance > 0
    
    def customize_board(self, matrix_board):
        for i in range(self.__n):
            for j in range(self.__m):
                state = matrix_board[i, j]
                self.__map_squares[(i, j)] = state
                self.__map_states[state] = (i, j)
                
    def init_R(self):
        size = self.get_size()
        R = zeros((size, size))
        
        for i in range(size):
            for j in range(size):
                if (self.is_neighbor(self.__map_states[i], self.__map_states[j])):
                    if (j == self.__goal_state):
                        R[i, j] = 100
                    else:
                        R[i, j] = 0
                elif (i == self.__goal_state and i == j):
                    R[i, j] = 100
                else:
                    R[i, j] = -100
        
        self.__R = R
````

### Clase QLearningAlgorithm

Esta clase implementa el algoritmo de entrenamiento Q-Learning, ``training``, a través del cual se actualiza la matriz Q, y que devuelve una lista con el rendimiento de cada uno de los episodios de entrenamiento.

La expresión que lleva a cabo dicha actualización es la siguiente:

```
Q[estado_actual, acción] = R[estado_actual, acción] + gamma * max(Q[estado_siguiente])
```

En esta clase se tienen los siguientes atributos:

- board: el tablero (Board) sobre el que se aplicará el algoritmo de entrenamiento.
- R: la matriz de recompensas.
- n_training: número de entrenamientos (episodios) que realizará el algoritmo.
- gamma: parámetro de aprendizaje.
- Q: matriz "cerebro" del agente.

Además del entrenamiento, esta clase también implementa la obtención del caminio desde la casilla inicial hasta la casilla objetivo, ``get_path``, una vez se tiene una matriz Q entrenada.

````python
class QLearningAlgorithm():
    def __init__(self, board, R, n_training, gamma):
        self.__board = board
        self.__R = R
        self.__n_training = n_training
        self.__gamma = gamma
        self.__Q = zeros((board.get_size(), board.get_size()))
        
    def get_Q(self):
        return self.__Q
    
    def get_n_training(self):
        return self.__n_training
    
    def training(self):
        goal_state = self.__board.get_goal_state()
        performances = []
        for i in range(self.__n_training):
            current_state = choice(list(range(self.__board.get_size())))
            while True:
                possible_actions = {}
                actions = self.__R[current_state]
                for i in range(self.__board.get_size()):
                    if (actions[i] >= 0):
                        possible_actions[i] = actions[i]
                next_state = choice(list(possible_actions.keys()))
                action = next_state                
                self.__Q[current_state, action] = self.__R[current_state, action] + self.__gamma * max(self.__Q[next_state])
                current_state = next_state
                if (current_state == goal_state):
                    break
            performances.append(self.__calculate_performance())
        self.__normalize_Q()
        return performances
    
    def get_path(self):
        map_states = self.__board.get_map_states()
        goal_state = self.__board.get_goal_state()
        current_state = self.__board.get_initial_state()
        path = [(current_state, map_states[current_state])]
        
        while (current_state != goal_state):
            actions = self.__Q[current_state]
            
            for i in range(0, len(path)):
                actions[path[i][0]] = 0
            
            max_val = max(actions)
            
            if (not max_val):
                msg.config(text="¡ME HE PERDIDO!\nCon esa configuración no consigo aprender correctamente.\nVuelve a intentarlo, por favor.")
                break
            
            current_state = choice(where(actions == max_val)[0])
            path.append((current_state, map_states[current_state]))
            
        return path
    
    def __calculate_performance(self):
        performance = 0
        
        for i in range(self.__board.get_size()):
            for j in range(self.__board.get_size()):
                performance = performance + self.__Q[i, j]
                
        max_value = self.__get_max_value()
        
        return (performance / max_value) * 100
    
    def __get_max_value(self):
        values = []
        for i in range(self.__board.get_size()):
            actions = self.__Q[i]
            values.append(max(actions))    
        return max(values)
    
    def __normalize_Q(self):
        max_value = self.__get_max_value()
        
        for i in range(self.__board.get_size()):
            for j in range(self.__board.get_size()):
                self.__Q[i, j] = round((self.__Q[i, j] / max_value) * 100)
````

### QLearningVariant Algorithm

Esta clase hereda de la clase anterior, con la particularidad de que introduce un par de parámetros nuevos:

- epsilon: parámetro con valor de 0 a 1.
- alpha: parámetro con valor de 0 a 1 que actúa como factor de descuento de epsilon.

En esta clase además se modifica la implementación del método ``training``. Esta variación del algoritmo pone en práctica la idea de "exploración" y "explotación".

El algoritmo genera en cada episodio de un entrenamiento un valor aleatorio k (entre 0 y 1) que se compara con epsilon.

- Si k es menor, se aplicará exploración, es decir, un comportamiento similar al del algoritmo de entrenamiento anterior.
- Si k es mayor, se aplicará explotación, esto es, se tomará la decisión de movimiento en base a la matriz Q, en base a lo que el agente ya sabe.

````python
class QLearningVariantAlgorithm(QLearningAlgorithm):
    def __init__(self, board, R, n_training, gamma, epsilon, alpha):
        super().__init__(board, R, n_training, gamma)
        self.__epsilon = epsilon
        self.__alpha = alpha

    def training(self):
        goal_state = self._QLearningAlgorithm__board.get_goal_state()
        performances = []
        for i in range(self._QLearningAlgorithm__n_training):
            current_state = choice(list(range(self._QLearningAlgorithm__board.get_size())))
            epsilon = self.__epsilon
            while True:
                possible_actions = {}
                actions = self._QLearningAlgorithm__R[current_state]
                for i in range(self._QLearningAlgorithm__board.get_size()):
                    if (actions[i] >= 0):
                        possible_actions[i] = actions[i]               
                k = random()
                if (k > epsilon):
                    max_val = max(self._QLearningAlgorithm__Q[current_state])
                    next_state = choice(where(self._QLearningAlgorithm__Q[current_state] == max_val)[0])
                else:
                    next_state = choice(list(possible_actions.keys()))
                epsilon = epsilon * self.__alpha
                action = next_state
                self._QLearningAlgorithm__Q[current_state, action] = self._QLearningAlgorithm__R[current_state, action] + self._QLearningAlgorithm__gamma * max(self._QLearningAlgorithm__Q[next_state])
                current_state = next_state
                if (current_state == goal_state):
                    break
            performances.append(self._QLearningAlgorithm__calculate_performance())
        self._QLearningAlgorithm__normalize_Q()
        return performances
````

### Funciones auxiliares

Mediante estas dos funciones se consigue las siguientes operaciones:

- ``get_graph(board)``: construcción de una imagen que representa el tablero que se pasa como parámetro.
- ``get_graphpath(board, path)``: construcción de una imagen que representa el tablero y el camino que resulta de la secuencia de casillas que une la casilla inicial con la casilla objetivo.

````python
def get_graph(board):
    dot = graphviz.Digraph(engine="neato", format="png")
    nodes = list(board.get_map_states().keys())
    
    for i in range(0, board.get_size()):
        dot.node(str(nodes[i]), pos=str(board.get_map_states()[nodes[i]][1]) + ",-" + str(board.get_map_states()[nodes[i]][0]) + "!")
        
    dot.render('InitBoard', view=False)

def get_graphpath(board, path):
    dot = graphviz.Digraph(engine="neato", format="png")
    nodes = list(board.get_map_states().keys())

    for i in range(0, board.get_size()):
        dot.node(str(nodes[i]), pos=str(board.get_map_states()[nodes[i]][1]) + ",-" + str(board.get_map_states()[nodes[i]][0]) + "!")
    
    for i in range(0, len(path)):
        if i+1 == len(path):
            break
        else:
            dot.edge(str(path[i][0]), str(path[i+1][0]))
    
    dot.render('PathBoard', view=False)
````
